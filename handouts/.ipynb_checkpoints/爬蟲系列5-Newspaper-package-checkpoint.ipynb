* Text extraction from html
* Top image extraction from html

* Keyword extraction from text
* Summary extraction from text
* Author extraction from text
* Google trending terms extraction
* Works in 10+ languages (English, Chinese, German, Arabic, …)


# Let's see what newspaper can do!


```python
import newspaper
from newspaper import Article
```

```python
# Interact ipython
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
newspaper.languages()
```

```python
cnn_paper = newspaper.build('https://udn.com')
#for article in cnn_paper.articles:
#    print (article.url)
for category in cnn_paper.category_urls():
    print(category)
```

```python
# English example
url = "https://cruxnow.com/vatican/2018/02/07/benedict-xvi-writes-pilgrimage-towards-home-last-tiring-stage-journey/"
# 中文例子
url2 = "https://udn.com/news/story/7269/3069078?from=udn-catebreaknews_ch2"
article = Article(url2,language='zh')
article.download()
article.html
```

```python
article.parse()
article.authors
#article.publish_date
article.text
```

```python
article.nlp('')
article.keywords
article.summary
```
