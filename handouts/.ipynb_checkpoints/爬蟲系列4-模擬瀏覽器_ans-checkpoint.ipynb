````python
# 使用 selenium 的定位功能
con = driver.find_elements_by_class_name("value")  #抓取有 class = value 的 element(Tag)
con_text = [item.text for item in con]   #取得 element 下的文字

con_text
````

# 如果想要找到特定路徑的 tag 呢?  => XPath
當碰到有些網頁較無法透過 class 或 id 去定位時，XPath 會較適合，但是如果能透過標籤與其屬性來定位時，就盡量採用，防止只要網站稍有變動就抓不到資料。

````python
# XPath 介紹
items1 = driver.find_elements_by_xpath("//*[@id='HEADER']/div/div[2]/a[1]")
items1[0].text

items2 = driver.find_elements_by_xpath("//*[@id='Block12Container7']/div/div/h5/a/font/strong")
items2[0].text
````

## 再一次回顧上面例子的不同
在這種需要前端瀏覽器即時加載js的網頁，我們最好的方式就是用模擬瀏覽器的方式去模擬，等確定載入網頁後，再取得網頁原始碼。<br>
### 然而，究竟要等多久呢?

1. 強制等待
```
from time import sleep
sleep(3)
```
2. 隱式等待 <br>
告訴 webdriver 我最長等多久的時間，不然就繼續執行
`driver.implicitly_wait()`


````python
# 強制等待
from time import sleep
print("hi")
sleep(3)
print("3 sec")
````

````python
# 隱式等待
from selenium import webdriver
driver = webdriver.Chrome(driver_path)
driver.maximize_window() 
driver.implicitly_wait(10) # seconds
driver.get('http://24h.pchome.com.tw/region/DHBE')
````

###### 3.判斷條件等待<br>
**`WebDriverWait(driver, 超時時長, 调用频率, 忽略異常).until(可執行方法, 超時時返回的信息)`** <br>
可使用的判斷條件
    * title_is
    * title_contains
    * presence_of_element_located
    * visibility_of_element_located
    * visibility_of
    * presence_of_all_elements_located
    * text_to_be_present_in_element
    * text_to_be_present_in_element_value
    * frame_to_be_available_and_switch_to_it
    * invisibility_of_element_located
    * element_to_be_clickable
    * staleness_of
    * element_to_be_selected
    * element_located_to_be_selected
    * element_selection_state_to_be
    * element_located_selection_state_to_be
    * alert_is_present* 

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome(driver_path)
driver.get('http://24h.pchome.com.tw/region/DHBE')
try:
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "value"))
    )
    if (element):
        print("Get it!")
    else:
        print("Cannot reach the element!")
    
finally:
    driver.quit()
````
